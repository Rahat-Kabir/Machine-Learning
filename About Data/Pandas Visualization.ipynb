```python
import pandas as pd
df = pd.read_csv("iris.csv")
```

```python
df.shape
```

```python
df.info()
```

# Data Visualization

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

# Count Plot

```python
sns.countplot(data=df,y='species')
```

```python
sns.barplot(data=df,x='species',y='sepal_length')
plt.show
```

```python
sns.swarmplot(data=df,x='species',y='petal_length') 
```

```python
#set the plot fig size
plt.figure(figsize=(20,10))

#count plot
sns.countplot(data=df,x='petal_length',hue='species')

#labeling
plt.ylabel('Count per Species')
plt.title('Iris Flower Species Analysis')

#show plot
plt.show()
```

# Box plot

```python
sns.boxplot(data=df,x='species',y='sepal_length')
```

# Pair Plot

```python
sns.pairplot(df,hue='species')
plt.show()
```

# His plot

```python
sns.histplot(data=df)
```

```python
sns.histplot(data=df, x="sepal_length")
plt.show()
```

```python
sns.histplot(data=df, x="sepal_length")
plt.grid()
plt.show()
```

```python
sns.histplot(data=df, x="sepal_length", log_scale=True, element="step", fill=False)
```

```python
#histograms together
df.hist(edgecolor='black',linewidth=1.2)
plt.tight_layout()
plt.show
```

# Distributation plot

```python
sns.distplot(df.petal_length)
```

```python
sns.scatterplot(data=df)
```
